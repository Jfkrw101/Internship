```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_pose = pd.read_csv("../assets/results.csv")
```

```python
df_pose.head()
```

```python
df_pose['           val/cls_loss']
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_pose['                  epoch'], df_pose['         train/box_loss'], label='Train Box Loss', linestyle='-', color='blue')
plt.plot(df_pose['                  epoch'], df_pose['           val/box_loss'], label='Val Box Loss', linestyle='--', color='blue')

plt.plot(df_pose['                  epoch'], df_pose['         train/cls_loss'], label='Train Class Loss', linestyle='-', color='orange')
plt.plot(df_pose['                  epoch'], df_pose['           val/cls_loss'], label='Val Class Loss', linestyle='--', color='orange')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.show()
```

```python
df_pose["       metrics/mAP50(B)"]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_pose['                  epoch'], df_pose['   metrics/precision(B)'], label='Precision (B)', linestyle='-', color='green')
plt.plot(df_pose['                  epoch'], df_pose['      metrics/recall(B)'], label='Recall (B)', linestyle='-', color='red')
plt.plot(df_pose['                  epoch'], df_pose["       metrics/mAP50(B)"], label='mAP50 (B)', linestyle='-', color='purple')

plt.xlabel('Epochs')
plt.ylabel('Metric Score')
plt.title('Precision, Recall, and mAP50')
plt.legend()
plt.show()
```
